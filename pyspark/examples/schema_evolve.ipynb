```
from pyspark.sql.functions import lit
```

```
df1 = spark.range(3)
df1.write.format("delta").mode("overwrite").save("/tmp/delta/customer2")
```

```
df1.show()
```

```
df2 = spark.range(2).withColumn("customer_id", lit(2))
df2.write.format("delta").option("mergeSchema", True).mode("append").save("/tmp/delta/customer2")
```

```
df2.show()
```

```
spark.read.format("delta").load("/tmp/delta/customer2").show()
```
